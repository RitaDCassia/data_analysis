```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# criando uma varuavel com dados de uma distribuicão nornal com a
# funcao rvs (100 elementos)
dados = norm.rvs(size = 100)
dados
```

```python
# histograma
# bins é o numero de barras
plt.hist(dados, bins = 20)
plt.title("Dados")
```

```python
# geracao de grafico pra verificar se a distribuicao é normal
fig, ax = plt.subplots()
# fazemos a impressao do grafico
stats.probplot(dados, fit = True, plot=ax)
plt.show()
```

```python
# teste de hipotese
# segundo argumento é um valor de p, não há como rejeitar a hipótese nula
stats.shapiro(dados)
```

```python
# dados nao normais
# gerando 1000 dados enviezados
dados2 = skewnorm.rvs(4, size=1000)
dados2
```

```python
# histogtama
plt.hist(dados2, bins=20)
plt.title("Dados")
```

```python
# gracao de graficos sem distribuicao normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
