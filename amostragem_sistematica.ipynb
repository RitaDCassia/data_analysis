```python
import numpy as np
import pandas as pd
from math import ceil
```

```python
# criacao de variaveis para representar a populacao, amorstra e o valor de k
populacao = 150
amostra = 15
# ceil -> maior valor do resultado de uma divisao um numero real
k = ceil(populacao / amostra)
print(k)
```

ira extrair um elemento a cada dez

```python
# definicao do valor randomico para inicializar a amostra, inicando em 1 ate k + 1
# menor valor 1 maior valor k
r = np.random.randint(low = 1, high= k + 1, size =1)
print(r)
```

```python
# for para somar os proximos valores, baseado no primeiro valor r que foi definido acima
acumulador = r[0]
sorteados = []

for i in range(amostra):
  sorteados.append(acumulador)
  acumulador += k

print(sorteados)
```

```python
# carregamos a base de dados e criamos a base_final somemnte com valores sorteados
base = pd.read_csv('iris.csv')
# usa os elementos nos registros
base_final = base.loc[sorteados]
base_final
```
