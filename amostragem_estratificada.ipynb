```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
iris = pd.read_csv('iris.csv')
iris['class'].value_counts()
```

```python
iris
```

```python
# klearn.model_selection usada para dividir matrizes ou dataframes em subconjuntos aleatÃ³rios de treino e teste.
# iris.iloc[:, 0:4] -> busca apenas os atributos previsores, os dados sobre a pÃ©tala
# e a sÃ©tala da planta
# iris.iloc[:, 4] -> Busca domente a classe que Ã© a especie da planta (setosa, virginica ou visicolor)
# test_size=0.5 -> selecionamos 50% da base de dados, que serao copiados para as
# variaveis x e y, essa funcao retorna 4 valores. porÃ©m vamos usar domente os 50%
# da base de dados e por isso colocamos "_" para os outros valores
# stratify= iris.iloc[:, 4] -> para tornar a amostra baseadad na classe, amostra estratidicada da coluna 4
x, _, y, _, = train_test_split(iris.iloc[:, 0:4], iris.iloc[:, 4],
                               test_size=0.5, stratify= iris.iloc[:, 4])

y.value_counts()
```
